```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_m1 = pd.read_csv('apple_m1_pro_macbook.csv')
df_i7_8750 = pd.read_csv('intel_i7_8750H_macbook.csv')
```

```python
base_frequency = 2600

plt.figure(figsize=(30, 10))

sns.set_style("whitegrid")
ax = sns.lineplot(x=df_i7_8750.index, y='CPU Requsted Frequency_0(MHz)', data=df_i7_8750, linewidth=0.5, label='CPU frequency')
ax.axhline(y=base_frequency, color='blue', linestyle='--', label='Maximum cpu frequency')

ax.set(xlabel='Sample points', ylabel='CPU Frequency (MHz)', title='CPU Frequency Through Time (i7 8750H, Macbook Pro 15)')
ax.legend()
sns.set_palette('colorblind')

plt.show()
```

```python
# Set base frequency value
base_frequency_e_core = 2064
base_frequency_p_core = 3220

# Create the figure and the first line plot with Seaborn
fig, ax1 = plt.subplots(figsize=(30, 10))
ax1 = sns.lineplot(x=df_m1.index, y=' E-Core Frequency (Mhz)', data=df_m1, linewidth=0.5, color='blue', label='E-core Frequency')

# Create a second y-axis and the second line plot
ax1 = sns.lineplot(x=df_m1.index, y=' P-Core Frequency (Mhz)', data=df_m1, linewidth=0.5, color='orange', label='P-Core Frequency')

# Add the base frequency line
ax1.axhline(y=base_frequency_e_core, color='green', linestyle='--', label='E-Core maximum frequency')
ax1.axhline(y=base_frequency_p_core, color='red', linestyle='--', label='P-Core maximum frequency')

# Set the x-axis label and title
ax1.set(xlabel='Sample points', ylabel='CPU Frequency (MHz)', title='CPU Frequency Through Time (M1 Pro, Macbook Pro 14)')

ax1.legend()

# Show the plot
plt.show()
```

```python
df_i5_1235u_turbo_on = pd.read_csv('i5_1235u_turbo_on.csv')
df_i5_1235u_turbo_off = pd.read_csv('i5_1235u_turbo_off.csv')
```

```python
df_i5_1235u_turbo_off = df_i5_1235u_turbo_off.drop(df_i5_1235u_turbo_off.tail(1307).index)
```

```python
# Select every 10th row starting from the first row (index 0)
sampled_data = df_i5_1235u_turbo_on.iloc[::10, :]

# Reset the index of the sampled_data DataFrame
sampled_data.reset_index(drop=True, inplace=True)
```

```python
sample_data_off = df_i5_1235u_turbo_off.iloc[::10, :]
sample_data_off.reset_index(drop=True, inplace=True)
```

```python
# Set base frequency value
base_frequency = 1300
base_frequency_turbo = 4400

# Create the figure and the first line plot with Seaborn
fig, ax1 = plt.subplots(figsize=(30, 10))
ax1 = sns.lineplot(x=sample_data_off.index, y='CPU Frequency_0(MHz)', data=sample_data_off, linewidth=0.5, color='blue', label='Turbo-boost off')

# Create a second y-axis and the second line plot
ax1 = sns.lineplot(x=sampled_data.index, y='CPU Frequency_0(MHz)', data=sampled_data, linewidth=0.5, color='orange', label='Turbo-boost on')

# Add the base frequency line
ax1.axhline(y=base_frequency, color='green', linestyle='--', label='Base frequency')
ax1.axhline(y=base_frequency_turbo, color='red', linestyle='--', label='Turbo-boost maximum frequency')

# Set the x-axis label and title
ax1.set(xlabel='Sample points', ylabel='CPU Frequency (MHz)', title='CPU Frequency Through Time (Dell Inspirtion 14, i5 1235U)')

ax1.legend()

# Show the plot
plt.show()
```

```python
# Set base frequency value
base_frequency = 100

# Create the figure and the first line plot with Seaborn
fig, ax1 = plt.subplots(figsize=(30, 10))
ax1 = sns.lineplot(x=sample_data_off.index, y=' CPU Utilization(%)', data=sample_data_off, linewidth=0.5, color='blue', label='Turbo-boost off')

# Create a second y-axis and the second line plot
ax1 = sns.lineplot(x=sampled_data.index, y=' CPU Utilization(%)', data=sampled_data, linewidth=0.5, color='orange', label='Turbo-boost on')

# Add the base frequency line
ax1.axhline(y=base_frequency, color='green', linestyle='--', label='Maximum utility for a single core')

# Set the x-axis label and title
ax1.set(xlabel='Sample points', ylabel='CPU Utilization(%)', title='CPU utilization Through Time (Dell Inspirtion 14, i5 1235U)')

ax1.legend()

# Show the plot
plt.show()
```

```python
# Create the figure and the first line plot with Seaborn
fig, ax1 = plt.subplots(figsize=(30, 10))
ax1 = sns.lineplot(x=sample_data_off.index, y='Cumulative Processor Energy_0(mWh)', data=sample_data_off, linewidth=0.5, color='blue', label='Turbo-boost off')

# Create a second y-axis and the second line plot
ax1 = sns.lineplot(x=sampled_data.index, y='Cumulative Processor Energy_0(mWh)', data=sampled_data, linewidth=0.5, color='orange', label='Turbo-boost on')

# Set the x-axis label and title
ax1.set(xlabel='Sample points', ylabel='CPU Power consumption (mWh)', title='CPU Power consumption Through Time (Dell Inspirtion 14, i5 1235U)')

ax1.legend()

# Show the plot
plt.show()
```
